```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add edges representing the food chain
G.add_edge("Plants", "Grasshoppers")
G.add_edge("Grasshoppers", "Frogs")
G.add_edge("Grasshoppers", "Birds")
G.add_edge("Frogs", "Snakes")
G.add_edge("Frogs", "Birds")
G.add_edge("a", "l")

# Draw the graph
pos = nx.spring_layout(G)  # use Spring layout for a better appearance
nx.draw(G, pos, with_labels=True, node_size=500, node_color="lightblue", font_size=10, font_weight="bold", edge_color="gray")
plt.title("Sample Food Web")
plt.show()
```
